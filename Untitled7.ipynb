```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

```

```python
x1 = np.array([30,30,40, 40])
y1 = np.array([37, 47, 50, 60])
model = LinearRegression()
x= x1.reshape(-1, 1)
model.fit(x, y1)
print(x)
print(model.intercept_)
print(model.coef_)
plt.scatter(x1, y1)
plt.plot(x, 3+1.3*x)
model.score(x, y1)
```

```python
x2 = np.array([30,30,40, 40, 20, 20, 50, 50])
y2 = np.array([37, 47, 50, 60, 25, 35, 62, 72])
model = LinearRegression()
x= x2.reshape(-1, 1)
model.fit(x, y2)
print(x)
print(model.intercept_)
print(model.coef_)
plt.scatter(x2, y2)
plt.plot(x, 5+1.24*x)
model.score(x, y2)
```

```python
x3 = np.array([30,30,40, 40, 20, 20, 50, 50, 10, 10, 60, 60])
y3 = np.array([37, 47, 50, 60, 25, 35, 62, 72, 13, 23, 74, 84])
model = LinearRegression()
x= x3.reshape(-1, 1)
model.fit(x, y3)
print(x)
print(model.intercept_)
print(model.coef_)
plt.scatter(x3, y3)
plt.plot(x, 5.6+1.225*x)
model.score(x, y3)
```

```python
x1 = np.array([10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5])
y1 = np.array([8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68])
model = LinearRegression()
x= x1.reshape(-1, 1)
model.fit(x, y1)
print(x)
print(model.intercept_)
print(model.coef_)
plt.scatter(x1, y1)
plt.plot(x, 3+0.5*x)
predict = 3 + 0.5*x
model.score(x, y1)
ost = x - predict
print(ost)
```
