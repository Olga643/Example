## Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr().round(2)
corr_matrix[np.abs(corr_matrix)<0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(20,10))
sns.set(font_scale=1.4)
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')
```

Матрица корреляции показывает, что на цену дома самое существенное влияние оказывает количество квадратных метров жилой площади, далее по убыванию идут конструкция здания и дизайн, квадратные метры дома, которые находятся выше земли, как ни странно - квадратные метры жилой площади у 15 соседей (видимо этот показатель подчеркивает статус района), незначительно влияют количество ванных комнат и вид
