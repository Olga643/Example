### 1.1. Скачать следующие данные: kc-house-data

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('kc_house_data.csv', encoding='1251', index_col='id', parse_dates=['date'], dayfirst=False )
df.head()
```

```python
df['price'].max(), df['price'].min(), df['price'].mean(), df['price'].median()
```

```python
df['price'].value_counts()
```

### 1.2. Изучите стоимости недвижимости

```python
plt.figure(figsize=(10, 7))
plt.hist(df['price'], bins = 50)
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Количество')
plt.grid()
```

```python
plt.figure(figsize=(10, 7))
sns.histplot(df['price'], bins = 50);
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Количество')
plt.grid()
```

```python
plt.figure(figsize=(20, 10))

sns.boxplot(x=df['price'])
plt.title('Распределение стоимости недвижимости')
plt.grid();
```

### 1.3. Изучите распределение квадратуры жилой площади

```python
df['sqft_living'].value_counts()
```

```python
df['sqft_living'].max(), df['sqft_living'].min(), df['sqft_living'].mean(), df['sqft_living'].median()
```

```python
plt.figure(figsize=(10, 7))
plt.hist(df['sqft_living'], bins = 50)
plt.title('Распределение квадратуры жилой площади')
plt.xlabel('Кв. метры жилые')
plt.ylabel('Количество')
plt.grid()
```

```python
plt.figure(figsize=(10, 7))

sns.boxplot(x=df['sqft_living'])
plt.title('Распределение квадратуры жилой площади')
plt.grid();
```
