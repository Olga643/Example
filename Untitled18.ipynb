```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('laptop_price.csv', encoding='1251', index_col='laptop_ID')
df.head()
```

### 1.1 Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU (Intel, AMD и т.д)

```python
df['Cpu'].value_counts()
```

```python
df['Cpu_Company'] = df['Cpu'].apply(lambda row: row.split()[0])
df.sample(10)
```

### 1.2 Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя (128, 1024, 256 и т.д.

```python
df['Memory'].value_counts()
```

512GB SSD +  1.0TB Hybrid

```python
def mem(row):
  gb = 0
  new_row = row.split()
  for elem in new_row:
    if 'GB' in elem:
      gb += int(elem[:-2])
    elif 'TB' in elem:
      gb += float(elem[:-2])*1024
  return int(gb)

df['Memory_Amount'] = df['Memory'].apply(mem)
df.sample()
```

### 1.3 Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD, SDD и т.д.)

```python
def mem_type(row):
  mem_row = []
  row_new = row.split()
  for  el in row_new:
    if 'GB' not in el and 'TB' not in el and el not in mem_row:
      mem_row.append(el)


  if len(mem_row) == 2 and mem_row[1] == '+':
    mem_row = [mem_row[0]]
  return ' '.join(mem_row)

df['Memory_Type'] = df['Memory'].apply(mem_type)
df.sample(10)
```

### 1.4 Удалите признаки Memory и ScreenResolution

```python
df.drop(columns=['Memory', 'ScreenResolution'], inplace=True)
df.sample(5)
```
