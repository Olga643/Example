Даны значения величины заработной платы заемщиков банка (zp) и значения их
поведенческого кредитного скоринга (ks): zp = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110],
ks = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832]. Используя математические
операции, посчитать коэффициенты линейной регрессии, приняв за X заработную плату
(то есть, zp - признак), а за y - значения скорингового балла (то есть, ks - целевая
переменная). Произвести расчет как с использованием intercept, так и без.

1) вариант с интерсептом

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
y = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
plt.scatter(x, y)
```

На графике прослеживается линейная зависимость, значит можно построить модель линейной регрессии

```python
model = LinearRegression()
x1 = x.reshape(-1, 1)
model.fit(x1, y)
print(model.intercept_)
print(model.coef_)
```

```python
n = len(x)
b1 = (n*np.sum(x*y) - np.sum(x) * np.sum(y))/(n*np.sum(x**2) - np.sum(x)**2)
b0 = np.mean(y) - b1*np.mean(x)
print(b0, b1)
```

```python
plt.scatter(x, y)
plt.plot(x, 444.177 + 2.62*x)
```

```python
model.score(x1, y)
```

```python
y_pred = 444.177 + 2.62 * x
print(y_pred)
mse = ((y - y_pred)**2).sum() / 10
print(mse)
```

2) вариант без интерсепта

```python
x_x = x.reshape(1, 10)
y_y = y.reshape(1, 10)
b_1 = np.dot(np.dot(np.linalg.inv(np.dot(x_x, x_x.T)), x_x), y_y.T)
print(b_1)
```

Задача 2 Посчитать коэффициент линейной регрессии при заработной плате (zp), используя
градиентный спуск (без intercept).

```python
def mse_(B1, y = y, x = x, n = len(x)):
  return np.sum((B1*x - y)**2) / n

alpha = 0.00001
B1 = 0.1
n = len(x)

for i in range (10000):
  B1 -= alpha * (2/n) * np.sum((B1 * x - y) * x)
  if i % 500 == 0:
    print('Iteration = {i}, B1 = {B1}, mse = {mse}'.format( i = i, B1 = B1, mse = mse_(B1) )) 
```

Задача 3 (Дополнительно) Произвести вычисления как в пункте 2, но с вычислением intercept. Учесть, что
изменение коэффициентов должно производиться
на каждом шаге одновременно (то есть изменение одного коэффициента не должно
влиять на изменение другого во время одной итерации).

```python
def mse_(B0, B1, y = y, x = x, n = len(x)):
  return np.sum((B0 + B1*x - y)**2) / n

alpha = 0.00001
B1 = 0
B0 = 0
n = len(x)

for i in range (5000000):
  B0 -= alpha * (2/n) * np.sum((B0 + B1 * x - y))
  B1 -= alpha * (2/n) * np.sum((B0 + B1 * x - y) * x)
  if i % 1000000 == 0:
    print('Iteration = {i}, B0 = {B0}, B1 = {B1}, mse = {mse}'.format( i = i, B0 = B0, B1 = B1, mse = mse_(B0, B1) )) 
```
